```python
import matplotlib.pyplot as plt
a=["Biscuts","Flours","Chips","Shampoo"]
b=[100,250,150,50]
plt.plot(a,b,color="mediumslateblue",marker="h",linestyle="--")
plt.title("DAILY SALES REPORT")
plt.xlabel("ITEMS")
plt.ylabel("SALES")
plt.grid(True)
plt.show()
```

```python
plt.bar(a,b,color="mediumslateblue")
plt.title("STOCK REPORT")
plt.xlabel("STOCK")
plt.ylabel("SALES")
plt.show()
```

```python
import numpy as np
a=["Cadbury_DairyMilk", "Ferrero_Rocher", "KitKat", "Snickers", "LotteChocopie"]
b=np.array([60,40,80,50,150])
c=["chocolate","saddlebrown","sandybrown","peru","peachpuff"]
plt.pie(b,colors=c, labels=a, autopct="%1.1f%%", startangle=230)
plt.axis("equal")
plt.title("CHOCOLATE VARAITY")
plt.show()
```
